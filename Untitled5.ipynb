```python
import pandas as pd
import numpy as np
import seaborn as sns
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, SimpleRNN, Dense
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('/content/RNNDatasetDemo.xlsx - Sheet1.csv')
df
```

```python
X = np.array(df.iloc[:,:-1])
X = X.reshape(-1,4,1)
X.shape
```

```python
y = np.array(df.iloc[:,:-1])
y.shape
```

```python
print(X)
```

```python
df[['x1','x2']]
```

```python
df.loc[df.x1%2==0, ['x1','x2']]#to print even numbers
```

```python
from threading import active_count
model = Sequential()
model.add(SimpleRNN(64, activation='relu', input_shape=(4,1), return_sequences=True))#rnn layer
model.add(SimpleRNN(32, activation='relu', return_sequences=True))#rnn layer
model.add(SimpleRNN(16, activation='relu'))#it just a normal or feed forward layer
model.add(Dense(1,activation='relu'))
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
model.summary()
```

```python
model.fit(X,y, epochs=100, batch_size=16, verbose=1)
```

```python
testing_data = np.array([100, 104, 108, 112]).reshape(-1,4,1)
testing_data
```

```python
model.predict(testing_data)#to predict the next value
```
